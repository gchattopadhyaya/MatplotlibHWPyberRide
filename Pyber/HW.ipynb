```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

#Print the city table to understand the common column
city_df.head()
```

```python
#Print the ride table to understand the common column
ride_df.head()
```

```python
# Combine the data into a single dataset
# Merge the first two datasets on "Wrestler" so that no data is lost (should be 182 rows)
combined_cityride_df = pd.merge(city_df, ride_df,
                                 how='outer', on='city')

# Display the data table for preview
combined_cityride_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#X coordinate is Total Number of Rides per city
#Y coordinate is Average Fare per city.
# Build the scatter plots for each city types
# Incorporate the other graph properties

#Urban Scatter Plot
city_group_urban = combined_cityride_df.loc[combined_cityride_df["type"] == "Urban"].groupby([combined_cityride_df["city"]]) 
city_group_urban.head()
total_rides_per_city_urban = city_group_urban["city"].value_counts()
average_fare_per_city_urban = city_group_urban["fare"].mean()
driver_count_per_city_urban = city_group_urban["driver_count"].sum()

#Rural Scatter Plot
city_group_rural = combined_cityride_df.loc[combined_cityride_df["type"] == "Rural"].groupby([combined_cityride_df["city"]]) 
city_group_rural.head()
total_rides_per_city_rural = city_group_urban["city"].value_counts()
average_fare_per_city_rural = city_group_urban["fare"].mean()
driver_count_per_city_rural = city_group_urban["driver_count"].sum()

#Rural Scatter Plot
city_group_rural = combined_cityride_df.loc[combined_cityride_df["type"] == "Rural"].groupby([combined_cityride_df["city"]]) 
city_group_rural.head()
total_rides_per_city_rural = city_group_rural["city"].value_counts()
average_fare_per_city_rural = city_group_rural["fare"].mean()
driver_count_per_city_rural = city_group_rural["driver_count"].sum()


#Suburban Scatter Plot
city_group_sub = combined_cityride_df.loc[combined_cityride_df["type"] == "Suburban"].groupby([combined_cityride_df["city"]]) 
city_group_sub.head()
total_rides_per_city_sub = city_group_sub["city"].value_counts()
average_fare_per_city_sub = city_group_sub["fare"].mean()
driver_count_per_city_sub = city_group_sub["driver_count"].sum()


# Create a scatter plot which compares MPG to horsepower

plt.scatter(x=total_rides_per_city_urban, y=average_fare_per_city_urban, marker="o", facecolors="coral", edgecolors="black",
            s=driver_count_per_city_urban, alpha=0.90, label="Urban")
plt.scatter(x=total_rides_per_city_rural, y=average_fare_per_city_rural, marker="o", facecolors="gold", edgecolors="black",
            s=driver_count_per_city_rural, alpha=0.90, label="Rural")
plt.scatter(x=total_rides_per_city_sub, y=average_fare_per_city_sub, marker="o", facecolors="blue", edgecolors="black",
            s=driver_count_per_city_sub, alpha=0.90, label="Suburban")

plt.grid()

# Create a legend
# Incorporate a text label regarding circle size
# Save Figure
plt.title('Pyber Ride Sharing Data (2016)')
plt.legend(loc=1)
plt.xlabel("Total No Of Rides Per City")
plt.ylabel("Average Fare ($)")
plt.show()
plt.savefig("PyberRide.png")


```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
#Calculate Type Percents
Type = combined_cityride_df["type"].value_counts() 
Type_Percentage = (combined_cityride_df["type"].value_counts()/combined_cityride_df["type"].count()) * 100 
Type_Percentage

#Build Pie Chart
#Labels for the sections of our pie chart
labels = ["Urban", "Rural", "Suburban"]

#The values of each section of the pie chart
sizes = Type_Percentage

#The colors of each section of the pie chart
colors = ["coral", "lightskyblue", "gold"]

#Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=275)

plt.title('% of Total Fares Per City Type')
#Save Figure
plt.savefig("Type_Percentage.png")
```

```python
# Show Figure
plt.show()
```
